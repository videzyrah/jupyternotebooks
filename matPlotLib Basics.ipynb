```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

# Basic Line Plot

```python
x = [1,7,13,21,33,45]
y = np.cos(x)
z = np.sin(x)

#use dollar signs for LaTex equation fonts
plt.plot(x,y, linewidth = 2.0, label = '$cos(x)$')
plt.plot(x,z, linewidth = 4.0, label = '$sin(x)$')
#axis labels
plt.xlabel('x')
plt.ylabel('value')

plt.title('Line\nPlot')
#create legend
plt.legend()

```

# Scatter Plot

```python
x = [1,7,13,21,33,45]
y = [2,3,8,2,5,7]

x2 = [2,8,15,25,37,41]
y2 = [3,4,7,3,4,6]

plt.scatter(x,y, label = 'y', color = 'blue', marker = 'x')
plt.scatter(x2,y2, label = 'z', color = 'red', marker = '*', s=500)

#axis labels
plt.xlabel('x')
plt.ylabel('value')

plt.title('Scatter Plot')
#create legend
plt.legend()
```

# Pie Chart

```python
costs = [500,200,50,300]
labels = ['rent','utilities','phone', 'groceries']
colors = ['blue', 'yellow', 'red', 'green']


plt.pie(costs,
        labels = labels,
        colors = colors, 
        shadow = True, 
        explode=(0,0.1,0,0),
        autopct= '%1.1f%%')

plt.title('Budget Breakdown')
#create legend
plt.legend()
```

# Bar Chart

```python
x = [1,7,13,21,33,45]
y = [2,3,8,2,5,7]

x2 = [2,8,15,25,37,41]
y2 = [3,4,7,3,4,6]

plt.bar(x,y, label = 'Bars1', color = 'blue')
plt.bar(x2,y2, label = 'Bars2', color = 'red')
#axis labels
plt.xlabel('x')
plt.ylabel('value')

plt.title('Bar Chart')
#create legend
plt.legend()
```

# Histogram



```python
ages = [1,7,13,21,33,45,8,11,33,44,5,6,15,22,31]
bins = [0,10,20,30,40,50]

plt.hist(ages, bins, histtype='bar', rwidth=0.7)

plt.title('Age Distribution')
```

# Stack Plot

```python
quarters = [1,2,3,4]

rushing = [25,30,11,38]
passing = [18,90,35,75]

plt.figure()

plt.plot([],[], color = 'red', label = 'rushing', linewidth = 5)
plt.plot([],[], color = 'blue', label = 'passing', linewidth = 5)

plt.stackplot(quarters, rushing, passing, colors = ['blue','red'])

plt.xlabel('Quarter')
plt.ylabel('Yards')

plt.title('Total Yardage per Quarter')
#create legend
plt.legend()
```
