# Import Library Dependencies

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [4, 6]
```

# Assign files to read. Define conversion rates.

```python
csv_to_read='portfolio_9_22_2021.csv' #set new file here
excel_to_read='portfolio_9_22_2021.xlsx'#for currency type and asset class column generation, do not change
canadianX = 0.79
```

# Read files. Add Columns. Filter columns to view, sort list.
Remove .head() method to see complete list

```python
df =  pd.read_csv(csv_to_read)
xldf = pd.read_excel(excel_to_read)
currency_column = xldf["Trading Currency"]
asset_class_column = xldf["Asset Class"]
df['Trading Currency']= currency_column
df['Asset Class']= asset_class_column
df['Market Value'] = df['Market Value'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.filter(items=['Symbol', 'Position', 'Last','Market Value', 'Trading Currency', 'Asset Class' ]).sort_values('Symbol')
```

# Convert currencies to USD. Sort by asset class.

```python
df['Market Value']= np.where(df['Trading Currency']=='CAD', 
                           df['Market Value']*canadianX, 
                           df['Market Value'])
```

```python
df.filter(items=['Symbol', 'Position', 'Last','Market Value', 'Trading Currency', 'Asset Class']).sort_values(by='Asset Class')
```

# Calculating portfolio total.

```python
portfolioTotal = sum(df["Market Value"].values)
print ("Eric's Net worth: " + f'${portfolioTotal:,}')
```

# Asset class subtotal calculations

```python
df_gold = df[df["Asset Class"]=="Gold"]
gold_subtotal = sum(df_gold["Market Value"].values)

df_stock = df[df["Asset Class"]=="Stock"]
stock_subtotal = sum(df_stock["Market Value"].values)

df_bonds = df[df["Asset Class"]=="Bonds"]
bonds_subtotal = sum(df_bonds["Market Value"].values)

df_commodity = df[df["Asset Class"]=="Commodity"]
commodity_subtotal = sum(df_commodity["Market Value"].values)

df_volatility = df[df["Asset Class"]=="Volatility"]
volatility_subtotal = sum(df_volatility["Market Value"].values)

```

# Pie Chart Build

```python
computed_asset_array = [gold_subtotal, stock_subtotal, bonds_subtotal, commodity_subtotal, volatility_subtotal]
labels = ['Gold ' + f'${gold_subtotal:,}','Stock ' + f'${stock_subtotal:,}','Bonds ' + f'${bonds_subtotal:,}', 'Commodities ' + f'${commodity_subtotal:,}', 
          'Volatilities ' +  f'${volatility_subtotal:,}']
colors = ['yellow', 'blue', 'green', 'orange', 'red']

plt.gca().axis("equal")
pie= plt.pie(computed_asset_array,
        labels = labels,
        colors = colors, 
        shadow = True, 
        explode=(0,0.1,0,0,0),
        autopct= '%1.1f%%')

plt.title('Portfolio Distribution')
#create legend
plt.legend(pie[0],labels, bbox_to_anchor=(0,1), loc="right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
```

# Export Weekly Report as File

```python
computed_asset_array = [gold_subtotal, stock_subtotal, bonds_subtotal, commodity_subtotal, volatility_subtotal]
labels = ['Gold ' + f'${gold_subtotal:,}','Stock ' + f'${stock_subtotal:,}','Bonds ' + f'${bonds_subtotal:,}', 'Commodities ' + f'${commodity_subtotal:,}', 
          'Volatilities ' +  f'${volatility_subtotal:,}']
colors = ['yellow', 'blue', 'green', 'orange', 'red']

plt.gca().axis("equal")
pie= plt.pie(computed_asset_array,
        labels = labels,
        colors = colors, 
        shadow = True, 
        explode=(0,0.1,0,0,0),
        autopct= '%1.1f%%')
date = input("Enter report date: e.g. 01-01-2021 ")
plt.title('Portfolio Distribution for: ' + date)
#create legend
plt.legend(pie[0],labels, bbox_to_anchor=(0,1), loc="right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
report_output_filename = input("Enter file name for report export (use .png or .pdf extension")
plt.savefig(report_output_filename, bbox_inches="tight")
plt.show()
```
